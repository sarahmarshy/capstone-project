```python
import numpy as np
import pandas as pd
```

```python
#Database for area codes can be found here: http://www.nanpa.com/nanp1/npa_report.csv
#Excel file for database headings can be found here: http://www.nanpa.com/area_codes/AreaCodeDatabaseDefinitions.xls

#read in the the headings excel file for area code database
Area_Codes_heading = pd.read_excel('/home/calvin/Documents/PrivacyStar/data/AreaCodeDatabaseDefinitions.xls')

#only want the "Field name" column, which is our headings
Area_Codes_heading = Area_Codes_heading["Field Name"]

#assign the headings we got from the excel file to dataframe using the names parameter 
Area_Codes = pd.read_csv("/home/calvin/Documents/PrivacyStar/data/npa_report.csv", names=Area_Codes_heading)

#We need all the rows but columns NPA, Location, Country
#DataFrame.lock[rows, columns], ':' means that we want all the rows
Area_Codes = Area_Codes.loc[:,['NPA','Location', 'Country']]

#drops the first entry because its junk
Area_Codes.drop(labels=0, inplace=True)

Area_Codes.head(5)
```

```python
Phone_Record = pd.read_csv("/home/calvin/Documents/PrivacyStar/data/PhoneNumberTransactionsSeptember.txt", delimiter = "\t")
```

```python
External = pd.DataFrame()
User = pd.DataFrame()

#limit data size so local computer can handle it
Phone_Record = Phone_Record.head(1000)

#iterate through rows as tuples
for row in Phone_Record.itertuples():
    #construct two dataframes that contain (state, country) by crossreferencing NPA area codes and Phone_Record area codes
    External = External.append(Area_Codes[Area_Codes.NPA == str(row[3])], ignore_index=True)
    User = User.append(Area_Codes[Area_Codes.NPA == str(row[6])], ignore_index=True)

#Add four new columns to the Phone_Record Dataframe 
Phone_Record["ExternalState"] = External["Location"]
Phone_Record["ExternalCountry"] = External["Country"]
Phone_Record["UserState"] = User["Location"]
Phone_Record["UserCountry"] = User["Country"]
```
